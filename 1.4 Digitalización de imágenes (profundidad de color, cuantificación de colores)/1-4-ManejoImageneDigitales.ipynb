```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

image = 'lena_std.tif'
```

# Leyendo imagen

```python

# imagen en color
img = cv2.imread(image)
print('shape:', img.shape)
print('dtype:', img.dtype)

cv2_imshow(img)
```

```python
# imagen en escala de grises
img = cv2.imread(image, 0)
print('shape:', img.shape)
print('dtype:', img.dtype)

cv2_imshow(img)
```

# Redimensionar imagen

```python
# Redimension
print('original image shape:', img.shape)

width, height = 128, 256
resized_img = cv2.resize(img, (width, height))
print('resized to 128x256 image shape:', resized_img.shape)

cv2_imshow(resized_img)
```

```python
print('original image shape:', img.shape)
w_mult, h_mult = 0.25, 0.5
resized_img = cv2.resize(img, (0, 0), resized_img, w_mult, h_mult)
print('image shape:', resized_img.shape)

cv2_imshow(resized_img)
```

```python
print('original image shape:', img.shape)

w_mult, h_mult = 2, 4
resized_img = cv2.resize(img, (0, 0), resized_img, w_mult, 
                         h_mult, cv2.INTER_NEAREST)
print('image shape:', resized_img.shape)

cv2_imshow(resized_img)
```

# Rotar imagen

```python
img_flip_along_x = cv2.flip(img, 0)
img_flip_along_x_along_y = cv2.flip(img_flip_along_x, 1)
img_flipped_xy = cv2.flip(img, -1)

# cv2_imshow(img_flip_along_x)
cv2_imshow(img_flip_along_x_along_y)
# cv2_imshow(img_flipped_xy)
```

# Cambio formatos de imagen

```python
# save image with lower compression - bigger file size but faster decoding
cv2.imwrite("lena_png.png", img, [cv2.IMWRITE_PNG_COMPRESSION, 0])
# cv2.imwrite()
```

```python
# save image with lower quality - smaller file size
cv2.imwrite("lena_jpeg.jpeg", img, [cv2.IMWRITE_JPEG_QUALITY, 0])
```

# Dibujar en la imagen

```python
import random

image_circle = np.copy(img)
w, h = img.shape[1], img.shape[0]

def rand_pt(mult=1.):
    return (random.randrange(int(w*mult)),
            random.randrange(int(h*mult)))
```

```python
# Dibujando circulos
cv2.circle(image_circle, rand_pt(), 40, (255, 0, 0))
cv2.circle(image_circle, rand_pt(), 5, (255, 0, 0), cv2.FILLED)
cv2.circle(image_circle, rand_pt(), 40, (255, 85, 85), 2)
cv2.circle(image_circle, rand_pt(), 40, (255, 170, 170), 2, cv2.LINE_AA)

cv2_imshow(image_circle)
```

```python
# Dibujando lineas
image_line = np.copy(img)
cv2.line(image_line, rand_pt(), rand_pt(), (0, 255, 0))
cv2.line(image_line, rand_pt(), rand_pt(), (85, 255, 85), 3)
cv2.line(image_line, rand_pt(), rand_pt(), (170, 255, 170), 3, 
         cv2.LINE_AA)

cv2_imshow(image_line)
```

```python
# Dibujando mas elementos
image_elm = np.copy(img)

cv2.arrowedLine(image_elm, rand_pt(), rand_pt(), (0, 0, 255), 3, 
                cv2.LINE_AA)
cv2.rectangle(image_elm, rand_pt(), rand_pt(), (255, 255, 0), 3)
cv2.ellipse(image_elm, rand_pt(), rand_pt(0.3), random.randrange(360), 
            0, 360, (255, 255, 255), 3)

cv2_imshow(image_elm)
```

```python
# Colocando texto
image_txt = np.copy(img)

cv2.putText(image_txt, 'OpenCV', rand_pt(), 
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 3)

cv2_imshow(image_txt)
```
